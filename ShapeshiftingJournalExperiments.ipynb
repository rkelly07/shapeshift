# Naive Shapeshifting Algorithm

In this Notebook we will implement a naive solution for solving the shapeshifting problem. We do this by using one coordinator as the root of the new implementation and the other coordinator boat to move workers to different locations on the new configuration

## Methodology

To solve this problem we will take advantage of the geometries of the boats and how they are connected to find
the maximum overlap between our current configuration and our goal configuration at each step.

```python
init_config = gt.Graph(directed = True)
coordinator = init_config.add_vertex()
bottom_row = [init_config.add_vertex() for _ in range(size)]
top_row = [init_config.add_vertex() for _ in range(size)]

top_row_edges_forward = []
top_row_edges_backward = []

bottom_row_edges_forward = []
bottom_row_edges_backward = []

for i in range(len(bottom_row)-1):
    bottom_row_edges_forward.append(init_config.add_edge(bottom_row[i],bottom_row[i+1]))
    bottom_row_edges_backward.append(init_config.add_edge(bottom_row[i+1],bottom_row[i]))
    top_row_edges_forward.append(init_config.add_edge(top_row[i],top_row[i+1]))
    top_row_edges_backward.append(init_config.add_edge(top_row[i+1],top_row[i]))
    
top_bottom_edges = [init_config.add_edge(bottom_row[i],top_row[i]) for i in range(len(bottom_row))]
bottom_top_edges = [init_config.add_edge(top_row[i],bottom_row[i]) for i in range(len(bottom_row))]

init_edge_types = init_config.new_edge_property("int")

for i in range(len(top_row_edges_forward)):
    init_edge_types[top_row_edges_forward[i]] = 1
    init_edge_types[top_row_edges_backward[i]] =  2
    
    init_edge_types[bottom_row_edges_forward[i]] = 1
    init_edge_types[bottom_row_edges_backward[i]] = 2

for i in range(len(top_bottom_edges)):
    init_edge_types[top_bottom_edges[i]] = 3
    init_edge_types[bottom_top_edges[i]] = 4
    
coord_edge_right = init_config.add_edge(coordinator,top_row[0])
coord_edge_left = init_config.add_edge(top_row[0],coordinator)

init_edge_types[coord_edge_right] = 1
init_edge_types[coord_edge_left] = 2
init_config.edge_properties['connection_type'] = init_edge_types
```

```python
goal_config = gt.Graph(directed = True)
g_coordinator = goal_config.add_vertex()
left_column = [goal_config.add_vertex() for _ in range(size)]
right_column = [goal_config.add_vertex() for _ in range(size)]

left_column_edges_up = []
left_column_edges_down = []

right_column_edges_up = []
right_column_edges_down = []

for i in range(len(left_column)-1):
    left_column_edges_up.append(goal_config.add_edge(left_column[i],left_column[i+1]))
    left_column_edges_down.append(goal_config.add_edge(left_column[i+1],left_column[i]))
    right_column_edges_up.append(goal_config.add_edge(right_column[i],right_column[i+1]))
    right_column_edges_down.append(goal_config.add_edge(right_column[i+1],right_column[i]))

left_right_edges = [goal_config.add_edge(left_column[i],right_column[i]) for i in range(len(left_column))]
right_left_edges = [goal_config.add_edge(right_column[i],left_column[i]) for i in range(len(left_column))]

goal_edge_types = goal_config.new_edge_property("int")

for i in range(len(left_column_edges_up)):
    goal_edge_types[left_column_edges_up[i]] = 3
    goal_edge_types[left_column_edges_down[i]] = 4
    
    goal_edge_types[right_column_edges_up[i]] = 3
    goal_edge_types[right_column_edges_down[i]] = 4
    
for i in range(len(left_right_edges)):
    goal_edge_types[left_right_edges[i]] = 1
    goal_edge_types[right_left_edges[i]] = 2
    
g_coord_edge_right = goal_config.add_edge(g_coordinator,left_column[0])
g_coord_edge_left = goal_config.add_edge(left_column[0],g_coordinator)

goal_edge_types[g_coord_edge_right] = 1
goal_edge_types[g_coord_edge_left] = 2
goal_config.edge_properties['connection_type'] = goal_edge_types
```
